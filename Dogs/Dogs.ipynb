```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datascience/DogsandPuppies/dogs_dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Breed'].value_counts()
```

```python
df['Breed'].unique()
```

```python
plt.figure(figsize=(12,12))
breed_count = df['Breed'].value_counts()
breed_count.plot(kind='bar')
plt.title("Distribution of dog breeds")
plt.xlabel("Breed")
plt.ylabel("Count")
plt.show()
```

```python
coloured_dog = df['Color'].value_counts()
```

```python
coloured_dog.plot(kind='bar',color='yellow')
plt.title("Distribution of dog and its colour counts")
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=90, ha='right')
plt.grid(axis='y')
plt.show()
```

```python
gender_count = df['Gender'].value_counts()
gender_count.plot(kind='pie',autopct='%1.1f%%')
plt.title("Distribution of male and Female dogs")
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Breed',y='Weight (kg)',data=df)
plt.xticks(rotation=90,ha='right')
plt.xlabel("Breed")
plt.ylabel("Weight (kg)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Breed',y='Age (Years)',data=df)
plt.xticks(rotation=90,ha='right')
plt.xlabel("Breed")
plt.ylabel("Age average")
plt.show()
```
