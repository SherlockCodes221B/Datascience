```python
data = pd.read_csv('/kaggle/input/emoji-trends-dataset/emoji_usage_dataset.csv')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
platform_count = data['Platform'].value_counts()
platform_count.plot(kind='bar')
```

```python
gender_count = data['User Gender'].value_counts()
gender_count.plot(kind='pie',autopct='%1.1f%%')
plt.title("User Gender percentage")
plt.ylabel('')
plt.show()
```

```python
print("Age range max: ",data['User Age'].min(),"min: ",data['User Age'].max())
plt.title('Age Distribution Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
counts, bins, patches = plt.hist(data['User Age'], bins=10, edgecolor='black', color='skyblue')
```

```python
import warnings
warnings.filterwarnings("ignore")
sns.countplot(data=data,x=data['Emoji'])
```
